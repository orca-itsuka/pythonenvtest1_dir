### 練習問題2

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score
```

```python
target = iris["Species"]
data = iris.drop("Species", axis=1)
```

```python
LR = LogisticRegression(random_state=42,max_iter=200)
scores_LR = cross_val_score(LR,data,target)
print(scores_LR)
```

```python
Lsvc = LinearSVC(random_state=42,tol=1e-5,max_iter=10000)
scores_Lsvc = cross_val_score(Lsvc,data,target)
print(scores_Lsvc)
```

```python
svc = SVC(gamma="auto")
scores_svc = cross_val_score(svc,data,target)
print(scores_svc)
```

```python
DTC = DecisionTreeClassifier(random_state=42)
scores_DTC = cross_val_score(DTC,data,target)
print(scores_DTC)
```

```python
KNC = KNeighborsClassifier(n_neighbors=5)
scores_KNC = cross_val_score(KNC,data,target)
print(scores_KNC)
```

```python
Lreg = LinearRegression()
#scores_Lreg = cross_val_score(Lreg,data,target)
```

```python
SGD = SGDClassifier(random_state=42)
scores_SGD = cross_val_score(SGD,data,target)
print(scores_SGD)
```

```python
RFC = RandomForestClassifier(max_depth=2,random_state=42)
scores_RFC = cross_val_score(RFC,data,target)
print(scores_RFC)
```

```python
GBC = GradientBoostingClassifier(n_estimators=100,learning_rate=1.0,max_depth=1,random_state=42)
scores_GBC = cross_val_score(GBC,data,target)
print(scores_GBC)
```

```python
MLPC = MLPClassifier(random_state=42,max_iter=3000)
scores_MLPC = cross_val_score(MLPC,data,target)
print(scores_MLPC)
```

```python
KMS = KMeans(n_clusters=3, random_state=0, init='random')
scores_KMS = cross_val_score(KMS, data, target)
print(scores_KMS)
```
